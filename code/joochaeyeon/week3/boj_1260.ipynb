### DFS와 BFS
#### 문제 파악
- 그래프를 DFS로 탐색한 결과와 BFS로 탐색한 결과를 출력하는 프로그램을 작성하시오. 단, 방문할 수 있는 정점이 여러 개인 경우에는 정점 번호가 작은 것을 먼저 방문하고, 더 이상 방문할 수 있는 점이 없는 경우 종료한다. 정점 번호는 1번부터 N번까지이다.

#### 입력
첫째 줄에 정점의 개수 N(1 ≤ N ≤ 1,000), 간선의 개수 M(1 ≤ M ≤ 10,000), 탐색을 시작할 정점의 번호 V가 주어진다. 다음 M개의 줄에는 간선이 연결하는 두 정점의 번호가 주어진다. 어떤 두 정점 사이에 여러 개의 간선이 있을 수 있다. 입력으로 주어지는 간선은 양방향이다.

#### 출력
첫째 줄에 DFS를 수행한 결과를, 그 다음 줄에는 BFS를 수행한 결과를 출력한다. V부터 방문된 점을 순서대로 출력하면 된다.

#### 제한
- 시간제한 : 2초



```python
#정점의 개수 N, 간선의 개수 M, 탐색을 시작할 정점의 번호 V
N, M, V = map(int, input().split())

#그래프 초기화 -> 정점과 간선의 정보를 저장하는 인접행렬
graph = [[0] * (N + 1) for _ in range(N + 1)]

#다음 M개의 줄에는 간선이 연결하는 두 정점의 번호, 입력으로 주어지는 간선은 양방향
for i in range(M):
    a, b = map(int, input().split())
    graph[a][b] = graph[b][a] = 1  # 양방향이기에

#방문 리스트 행렬
#방문하지 않은 곳은 0으로 채워짐
visit1 = [0]*(N+1)
visit2 = [0]*(N+1)

def dfs(v): #깊이우선
    visit1[v]=1 #방문했다면 0->1로 변경
    print(v,end=" ")

    for i in range(1, N+1):
        if (graph[v][i] == 1 and visit1[i] == 0):
            dfs(i) #재귀처리

def bfs(v): #너비우선
    queue = [v]
    visit2[v]=1
    while queue:
        v = queue.pop(0)
        print(v,end=" ")
        for i in range(1,N+1):
            if (graph[v][i] == 1 and visit2[i]==0 ): 
                queue.append(i)
                visit2[i]=1
            
dfs(V)
print()
bfs(V)
print()
```
